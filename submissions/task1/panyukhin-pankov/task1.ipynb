```python
import matplotlib.pyplot as plt
from scipy.optimize import linprog;
import numpy as np;
 
def nash_equilibrium(inMatrix):
    inMatrix = np.array(inMatrix);
    mmin = inMatrix[inMatrix != 0].min();
    delta = abs(mmin) + 1;
    inMatrix = inMatrix + delta;
    transp = inMatrix.T*-1;
    c = np.ones((len(inMatrix)));
    bub = np.ones((len(inMatrix[0])))*-1;
    resz = linprog(c, transp, bub);
    resw = linprog(bub, inMatrix, c);
    v = 1 / sum(resz.x);
    return v * resz.x, v * resw.x, v - delta;
```

###  Пример 1: Существует равновесие Нэша.

```python
inMatrix = [[0, 4, 0],
            [3, 10, 1],
            [7, -6, 0],
            [-1, 6, 0]]
player_1, player_2, v = nash_equilibrium(inMatrix)
print('game value : ', v)
print('strategy of 1st player : ', player_1)
print('strategy of 2nd player : ', player_2)
```

```python
axisP = np.arange(1, int(len(inMatrix) + 1), 1)
axisQ = np.arange(1, len(inMatrix[0]) + 1, 1) 
plt.plot(axisP, player_1, 'ro')
plt.axis([0, len(inMatrix) + 2, 0, 2])
for i in range(len(inMatrix)):
    plt.plot([i+1, i+1], [0, player_1[i]])
plt.title('First player')
plt.show()
```

```python
plt.plot(axisQ, player_2, 'ro')
plt.axis([0, len(inMatrix[0]) + 2, 0, 2])
for i in range(len(inMatrix[0])):
    plt.plot([i+1, i+1], [0, player_2[i]])
plt.title('Second player')
plt.show()
```
